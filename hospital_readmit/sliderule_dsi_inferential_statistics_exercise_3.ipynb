# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

Null Hypothesis: The excess ratio for hospitals with less than or equal to100 discharges is the same as for >=1000 discharges.
Alternative hypothesis: The excess ratio for >1000 discharges is significantly higher.

```python
clean_hospital_read_df['Excess Readmission Ratio'].dropna()
```

```python
np.random.seed(42)

def ecdf(data):
    y=np.arange(1,len(data)+1)/len(data)
    x=np.sort(data)
    
    return x, y
cleandata=clean_hospital_read_df[['Number of Discharges','Excess Readmission Ratio']].dropna()
lessval=cleandata[cleandata['Number of Discharges']<=100]
moreval=cleandata[cleandata['Number of Discharges']>=1000]

x1,y1=ecdf(moreval['Excess Readmission Ratio'])
x2,y2=ecdf(lessval['Excess Readmission Ratio'])

# Plot the CDF with axis labels
_=plt.plot(x1,y1,marker='.',linestyle='none',color='blue')
_=plt.plot(x2,y2,marker='.',linestyle='none',color='red')
plt.show()
_=plt.hist(lessval['Excess Readmission Ratio'],normed=True)
_=plt.hist(moreval['Excess Readmission Ratio'],normed=True)
```

```python
lessval=cleandata[cleandata['Number of Discharges']<=100]
moreval=cleandata[cleandata['Number of Discharges']>=1000]
#observed difference in mean ratio:

obs_diff=lessval['Excess Readmission Ratio'].mean() - moreval['Excess Readmission Ratio'].mean()

newlist=pd.concat([lessval,moreval])




#bootstrap assuming mean is same for whole data set
np.random.seed(42)

def bootstrap_replicate_1d(data,func):
    bs_sample=np.random.choice(data,len(data))
    return func(bs_sample)

def draw_bs_reps(data, func, size=1):
    """Draw bootstrap replicates."""

    # Initialize array of replicates: bs_replicates
    bs_replicates = np.empty(size)

    # Generate replicates
    for i in range(size):
        bs_replicates[i] = bootstrap_replicate_1d(data,func)

    return bs_replicates

def diff_mean(data):
    diffs = np.mean(data[:int(len(data)/2)]) - np.mean( data[int(len(data)/2):] )
    return diffs

newar6=draw_bs_reps(newlist['Excess Readmission Ratio'],diff_mean,size=10000)
plt.hist(newar6, normed=True)
pval=np.sum(newar6>=obs_diff)/len(newar6)
confint=np.percentile(newar6,[0,99])
stanerr=np.std(newar6)
#margin of error, p value and confidence interval.
print("pval:",pval,", conf interval:",confint,", standard error:",stanerr,", observed difference",obs_diff)
```

ANSWER B<P>The result is statistically significant. There is a difference between the excess rations of large and small hospitals. In particular, large hospitals have a smaller ratio.

While our data is statistically significant, the practical significance is debatable. A difference in the rate of .04 is quite small. The cost to change hospital operations may not be justified if it produces such a small decrease in the ratio.
